```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
from scipy.stats import linregress
import gmaps
# Import API key
from api_keys import weather_api_key

# Import citipy to determine city based on latitude and longitude
from citipy import citipy
```

## Generate Cities List

```python
#create a list which will contain 500 cities around equator 

cities_coodinator = []
cities_name = []
equator_lat_range = (-90, 90)
equator_lng_range = (-180, 180)
#random 500 cities around equator
cities_lat = np.random.uniform(-90,90,643)
cities_lng = np.random.uniform(-180,180,643)
cities_coodinator = zip(cities_lat,cities_lng)

for coodinator in cities_coodinator:
    city = citipy.nearest_city(coodinator[0], coodinator[1]).city_name
    cities_name.append(city)
#     print(city)
# print(cities_name)
print(len(cities_name))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create a data frame

weather_dict = {
    
    "City": cities_name,
    "Lat" : '',
    "Lng" : '',
    "Max Temp" : '',
    "Humidity": '',
    "Cloudiness": '',
    "Wind Speed": '',
    "Country": '',
    "Date": ''

}

#Convert into DataFrame
weather_df = pd.DataFrame(weather_dict)

weather_df.head()
# weather_df.shape
```

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build query URL
query_url = f"{url}appid={weather_api_key}&unites={units}&q="
response  = requests.get(query_url+city).json()
pprint(response)
```

```python
# empty list
count = 1
cloud = []
new_cities =[]
country = []
date = []
humidity = []
temp = []
lats = []
longs =[]
wind =[]

print("Beginning Data Retrieval ")
print("-------------------------------")

# For loop matching city names with cities
for city in cities_name:
    
    try:
        response = requests.get(query_url+city).json()
        new_cities.append(response["name"])
        cloud.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        lats.append(response["coord"]["lat"])
        longs.append(response["coord"]["lon"])
        temp.append(response['main']['temp_max'])
        wind.append(response["wind"]["speed"])
        print(f"Processing Record {count} of Set {count} | {city}")
    
        #Increment row_counter
        count += 1
        
        #Pause program execution for 1 sec
        time.sleep(1.00)
        
    except(KeyError, IndexError):
        print("City not found. Skipping...")
        
        
print("-------------------------------")
print("Data Retrieval Complete")
print("-------------------------------")
```

## Linear Regression

```python
# Add the linear regression equation and line to plot
x_values = clean_city_data['Lat'].loc[clean_city_data['Lat']>= 0,]

y_values = clean_city_data['Max Temp'].loc[clean_city_data['Lat']>= 0,]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
print(f'The r-value is:{rvalue}')
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,fontsize=15,color="red",xy=(10, 270))
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

plt.show()
```
